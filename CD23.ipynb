```python
def infix_to_postfix(expr):
    prec = {'+':1, '-':1, '*':2, '/':2, '(':0}
    stack = []
    output = []

    for token in expr.split():
        # Operand (variable or number)
        if token.isalnum():
            output.append(token)

        # Left parenthesis
        elif token == '(':
            stack.append(token)

        # Right parenthesis
        elif token == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  # remove '('

        # Operator
        else:
            while stack and prec.get(stack[-1], 0) >= prec.get(token, 0):
                output.append(stack.pop())
            stack.append(token)

    # Pop remaining operators
    while stack:
        output.append(stack.pop())

    return " ".join(output)


# Test
expr = "price + discount * quantity"
print(infix_to_postfix(expr))
```
