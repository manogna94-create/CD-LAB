```python
# Step 1: Install (run once)
!pip install python-docx

# Step 2: Import
from docx import Document

# Step 3: Create a sample DOCX file
doc = Document()
doc.add_paragraph("This is the first line.")
doc.add_paragraph("This is the second line with more words.")
doc.add_paragraph("Microservices architecture example.")
doc.save("sample.docx")

# Step 4: Function
def analyze_docx(file_path):
    doc = Document(file_path)
    line_count = 0
    space_count = 0

    for para in doc.paragraphs:
        text = para.text
        if text.strip():
            line_count += 1
        space_count += text.count(' ')

    return line_count, space_count

# Step 5: Run
print(analyze_docx("sample.docx"))
```
