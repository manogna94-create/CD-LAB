```python
# Simple Lexer WITHOUT using PLY

import re

# Reserved words
reserved = {
    'ADD': 'ADD'
}

# Token patterns
token_specification = [
    ('NUMBER', r'\d+'),                       # Integer
    ('ID', r'[a-zA-Z_][a-zA-Z0-9_]*'),        # Identifier
    ('SKIP', r'[ \t]+'),                      # Skip spaces/tabs
    ('MISMATCH', r'.'),                       # Any other character
]

# Build regex
tok_regex = '|'.join('(?P<%s>%s)' % pair for pair in token_specification)

def lexer(code):
    for mo in re.finditer(tok_regex, code):
        kind = mo.lastgroup
        value = mo.group()

        if kind == 'NUMBER':
            value = int(value)
            yield ('NUMBER', value)

        elif kind == 'ID':
            kind = reserved.get(value, 'ID')
            yield (kind, value)

        elif kind == 'SKIP':
            continue

        elif kind == 'MISMATCH':
            print("Illegal character:", value)

# Input
data = "ADD item1 2"

# Run lexer
for token in lexer(data):
    print(token)
```
